```python
# Task 1: Top Cuisines Analysis
# Author: A.Jayasri
# Date: 11-02-26

import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('Dataset.csv')
print("Ready to analyze!")
```

```python
# Collect all cuisine types from the dataset
cuisine_collection = []
for item in data['Cuisines'].dropna():
    split_items = str(item).split(',')
    for food in split_items:
        cleaned_food = food.strip()
        cuisine_collection.append(cleaned_food)

print("Number of cuisine entries:", len(cuisine_collection))
```

```python
from collections import Counter
# Calculate frequency of each cuisine
cuisine_frequency = Counter(cuisine_collection)
# Get three most frequent cuisines
top_cuisines = cuisine_frequency.most_common(3)
print("Most popular cuisines:")
for rank, (food, freq) in enumerate(top_cuisines, start=1):
    percent = (freq / len(data)) * 100
    print(f"{rank}. {food} — {freq} entries ({percent:.2f}%)")
```

```python
from collections import Counter

cuisine_counts = Counter(cuisine_collection)
top_3 = cuisine_counts.most_common(3)
```

```python
# Prepare data for plotting
labels = []
values = []
for entry in top_3:
    labels.append(entry[0])
    values.append(entry[1])
# Draw bar chart
plt.figure(figsize=(9, 5))
bars = plt.bar(labels, values)
plt.xlabel("Cuisine Categories", fontsize=12)
plt.ylabel("Restaurant Count", fontsize=12)
plt.title("Top Three Popular Cuisines")
# Show numbers above bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             height,
             f"{int(height)}",
             ha='center',
             va='bottom')
plt.tight_layout()
plt.savefig('task1_top_cuisines.png')
plt.show()
print("Bar chart generated successfully")
```
